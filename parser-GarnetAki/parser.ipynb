```python
import requests
import html
import lxml
from tqdm import tqdm
from bs4 import BeautifulSoup as bs
import pandas as pd
import csv
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
import configparser

config = configparser.ConfigParser()
config.read("config.ini")

URL_TEMPLATE = config["URL"]["URL_TEMPLATE"]
URL_TEMPLATE_PLAYER = config["URL"]["URL_TEMPLATE_PLAYER"]
PATH_TO_SAVE_CSV = config["PATH"]["PATH_TO_SAVE_CSV"]
URL_STAT_PART = config["URL"]["URL_STAT_PART"]
GAME_CNT_IND = config["PARSE"]["GAME_CNT_IND"]
AVG_POINT_IND = config["PARSE"]["AVG_POINT_IND"]
CIV_WR_IND = config["PARSE"]["CIV_WR_IND"]
MAF_WR_IND = config["PARSE"]["MAF_WR_IND"]
SHER_WR_IND = config["PARSE"]["SHER_WR_IND"]
DON_WR_IND = config["PARSE"]["DON_WR_IND"]
GAME_IND_START = config["PARSE"]["GAME_IND_START"]
GAME_IND_END = config["PARSE"]["GAME_IND_END"]
CIV_WIN = config["PARSE"]["CIV_WIN"]
MAF_WIN = config["PARSE"]["MAF_WIN"]
PLAYERS_CNT = config["PARSE"]["PLAYERS_CNT"]

PATH_TO_SAVE_CSV='C://Users//1//Desktop//ML//MafiaData.csv'
URL_TEMPLATE='https://polemicagame.com/game-statistics/'
URL_TEMPLATE_PLAYER='https://polemicagame.com/cabinet/'
URL_STAT_PART='#statistics'
GAME_CNT_IND=1
AVG_POINT_IND=3
CIV_WR_IND=23
MAF_WR_IND=25
SHER_WR_IND=27
DON_WR_IND=29
GAME_IND_START=50000
GAME_IND_END=206380
CIV_WIN=1
MAF_WIN=2
PLAYERS_CNT=10

op = webdriver.ChromeOptions()
op.add_argument('headless')
driver = webdriver.Chrome(options=op)

d = dict()
```

```python
def get_civilian_info(player_ids, civilian_places):
    info = []
    for i in civilian_places:
        if d.get(player_ids[i]) == None:
            url = URL_TEMPLATE_PLAYER + str(player_ids[i]) + URL_STAT_PART
            driver.get(url)
            element = driver.find_element(By.CLASS_NAME, "stats-wrapper")
            str_from_element = element.text.splitlines()
            d[player_ids[i]] = [str_from_element[GAME_CNT_IND], str_from_element[AVG_POINT_IND], 
                                str_from_element[CIV_WR_IND].split(" ")[2], 
                                str_from_element[MAF_WR_IND].split(" ")[2], 
                                str_from_element[SHER_WR_IND].split(" ")[2], 
                                str_from_element[DON_WR_IND].split(" ")[2],]
            info.append(str_from_element[GAME_CNT_IND])
            info.append(str_from_element[AVG_POINT_IND])
            info.append(str_from_element[CIV_WR_IND].split(" ")[2])
        else:
            info.append(d.get(player_ids[i])[0])
            info.append(d.get(player_ids[i])[1])
            info.append(d.get(player_ids[i])[2])
    return info

def get_mafia_info(player_ids, mafia_places):
    info = []
    for i in mafia_places:
        if d.get(player_ids[i]) == None:
            url = URL_TEMPLATE_PLAYER + str(player_ids[i]) + URL_STAT_PART
            driver.get(url)
            element = driver.find_element(By.CLASS_NAME, "stats-wrapper")
            str_from_element = element.text.splitlines()
            d[player_ids[i]] = [str_from_element[GAME_CNT_IND], str_from_element[AVG_POINT_IND], 
                                str_from_element[CIV_WR_IND].split(" ")[2], 
                                str_from_element[MAF_WR_IND].split(" ")[2], 
                                str_from_element[SHER_WR_IND].split(" ")[2], 
                                str_from_element[DON_WR_IND].split(" ")[2],]
            info.append(str_from_element[GAME_CNT_IND])
            info.append(str_from_element[AVG_POINT_IND])
            info.append(str_from_element[MAF_WR_IND].split(" ")[2])
        else:
            info.append(d.get(player_ids[i])[0])
            info.append(d.get(player_ids[i])[1])
            info.append(d.get(player_ids[i])[3])
    return info

def get_sheriff_info(player_ids, sheriff_place):
    info = []
    if d.get(player_ids[sheriff_place]) == None:
        url = URL_TEMPLATE_PLAYER + str(player_ids[sheriff_place]) + URL_STAT_PART
        driver.get(url)
        element = driver.find_element(By.CLASS_NAME, "stats-wrapper")
        str_from_element = element.text.splitlines()
        d[player_ids[sheriff_place]] = [str_from_element[GAME_CNT_IND], str_from_element[AVG_POINT_IND], 
                                str_from_element[CIV_WR_IND].split(" ")[2], 
                                str_from_element[MAF_WR_IND].split(" ")[2], 
                                str_from_element[SHER_WR_IND].split(" ")[2], 
                                str_from_element[DON_WR_IND].split(" ")[2],]
        info.append(str_from_element[GAME_CNT_IND])
        info.append(str_from_element[AVG_POINT_IND])
        info.append(str_from_element[SHER_WR_IND].split(" ")[2]) 
    else:
            info.append(d.get(player_ids[sheriff_place])[0])
            info.append(d.get(player_ids[sheriff_place])[1])
            info.append(d.get(player_ids[sheriff_place])[4])
    return info

def get_don_info(player_ids, don_place):
    info = []
    if d.get(player_ids[don_place]) == None:
        url = URL_TEMPLATE_PLAYER + str(player_ids[don_place]) + URL_STAT_PART
        driver.get(url)
        element = driver.find_element(By.CLASS_NAME, "stats-wrapper")
        str_from_element = element.text.splitlines()
        d[player_ids[don_place]] = [str_from_element[GAME_CNT_IND], str_from_element[AVG_POINT_IND], 
                                str_from_element[CIV_WR_IND].split(" ")[2], 
                                str_from_element[MAF_WR_IND].split(" ")[2], 
                                str_from_element[SHER_WR_IND].split(" ")[2], 
                                str_from_element[DON_WR_IND].split(" ")[2],]
        info.append(str_from_element[GAME_CNT_IND])
        info.append(str_from_element[AVG_POINT_IND])
        info.append(str_from_element[DON_WR_IND].split(" ")[2])
    else: 
        info.append(d.get(player_ids[don_place])[0])
        info.append(d.get(player_ids[don_place])[1])
        info.append(d.get(player_ids[don_place])[5])
    return info
```

```python
def wr_append(game_info, civ_info, sher_info, maf_info, don_info):
    game_info.append("0." + civ_info[2][:-1])
    game_info.append("0." + civ_info[5][:-1])
    game_info.append("0." + civ_info[8][:-1])
    game_info.append("0." + civ_info[11][:-1])
    game_info.append("0." + civ_info[14][:-1])
    game_info.append("0." + civ_info[17][:-1])
    game_info.append("0." + sher_info[2][:-1])
    game_info.append("0." + maf_info[2][:-1])
    game_info.append("0." + maf_info[5][:-1])
    game_info.append("0." + don_info[2][:-1])
    return game_info

def games_cnt_append(game_info, civ_info, sher_info, maf_info, don_info):
    game_info.append(civ_info[0])
    game_info.append(civ_info[3])
    game_info.append(civ_info[6])
    game_info.append(civ_info[9])
    game_info.append(civ_info[12])
    game_info.append(civ_info[15])
    game_info.append(sher_info[0])
    game_info.append(maf_info[0])
    game_info.append(maf_info[3])
    game_info.append(don_info[0])
    return game_info

def avg_pts_append(game_info, civ_info, sher_info, maf_info, don_info):
    game_info.append(civ_info[1])
    game_info.append(civ_info[4])
    game_info.append(civ_info[7])
    game_info.append(civ_info[10])
    game_info.append(civ_info[13])
    game_info.append(civ_info[16])
    game_info.append(sher_info[1])
    game_info.append(maf_info[1])
    game_info.append(maf_info[4])
    game_info.append(don_info[1])
    return game_info
```

```python
def include_all_info(game_info, player_ids, civilian_places, mafia_places, sheriff_place, don_place):
    civ_info = get_civilian_info(player_ids, civilian_places)
    maf_info = get_mafia_info(player_ids, mafia_places)
    sher_info = get_sheriff_info(player_ids, sheriff_place)
    don_info = get_don_info(player_ids, don_place)
    game_info = wr_append(game_info, civ_info, sher_info, maf_info, don_info)
    game_info = games_cnt_append(game_info, civ_info, sher_info, maf_info, don_info)
    game_info = avg_pts_append(game_info, civ_info, sher_info, maf_info, don_info)

    return game_info
```

```python
def players_setup(game_info, json_data):
    player_ids, civilian_places, mafia_places, sheriff_place, don_place = [[], [], [], 0, 0]
    for i in range(0, 10, 1):
        player_ids.append(json_data['players'][i]['id'])
        if json_data['players'][i]['role']['type'] == 'civilian':
            civilian_places.append(i)
        if json_data['players'][i]['role']['type'] == 'mafia':
            mafia_places.append(i)
        if json_data['players'][i]['role']['type'] == 'sheriff':
            sheriff_place = i
        if json_data['players'][i]['role']['type'] == 'godfather':
            don_place = i
    game_info.append(sheriff_place+1)
    game_info.append(don_place+1)
    game_info.append(mafia_places[0]+1)
    game_info.append(mafia_places[1]+1)
    return include_all_info(game_info, player_ids, civilian_places, mafia_places, sheriff_place, don_place)
```

```python
def parse_json(json_data):
    game_info = []
    game_info.append(json_data['type'])
    if json_data['winnerCode'] == CIV_WIN:
        game_info.append('Civilian')
    elif json_data['winnerCode'] == MAF_WIN:
        game_info.append('Mafia')
    game_info.append(json_data['daysNumber'])
    return players_setup(game_info, json_data)
```

```python
def check_doctor_exists(json_data):
    if len(json_data['players']) != 10:
        return True
    
    for i in range(0, PLAYERS_CNT, 1):
        if json_data['players'][i]['role']['type'] == 'doctor':
            return True
    return False

def fill_data():
    data = []
    for ind in tqdm(range(GAME_IND_END, GAME_IND_START, -1)):
        try:
            session = requests.Session()
            response = session.get(URL_TEMPLATE + str(ind), allow_redirects=False)
            soup = bs(response.text, "html.parser")
            if soup.find("gamestats") is not None:
                game_data = soup.find("gamestats").get(':game-data')
                json_data = json.loads(game_data)
                if json_data['winnerCode'] != None:
                    if not check_doctor_exists(json_data):
                        data.append(parse_json(json_data))
        except:
            continue
    return data
```

```python
def create_csv():
    columns=['Type', 'Win', 'Days Number', 
             'Sheriff Place', 'Don Place', 'First Mafia Place', 'Last Mafia Place',
             '1 Civ WR', '2 Civ WR', '3 Civ WR', '4 Civ WR', '5 Civ WR', '6 Civ WR',
             'Sher WR', '1 Maf WR', '2 Maf WR', 'Don WR',
             '1 Civ games cnt', '2 Civ games cnt', '3 Civ games cnt', '4 Civ games cnt',
             '5 Civ games cnt', '6 Civ games cnt', 'Sher games cnt', '1 Maf games cnt',
             '2 Maf games cnt', 'Don games cnt', '1 Civ avg pts', '2 Civ avg pts',
             '3 Civ avg pts', '4 Civ avg pts', '5 Civ avg pts', '6 Civ avg pts',
             'Sher avg pts', '1 Maf avg pts', '2 Maf avg pts', 'Don avg pts']

    data = fill_data()
    
    df = pd.DataFrame(data, columns=columns)
    df.to_csv(PATH_TO_SAVE_CSV)
```
